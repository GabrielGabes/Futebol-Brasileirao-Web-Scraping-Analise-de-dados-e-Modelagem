```python
import pandas as pd

#Carregando os dados
df = pd.read_csv("jogo_bras.csv")
df
```

```python
# importando numero euler(e)
from math import e
print(e)

#Fatorial de um numero
def fatorial(numero):
    resultado=1
    count=1

    while count <= numero:
        resultado *= count
        count += 1
        
    return resultado

#Probabilidade Poisson
def dist_poisson(media, x):
    resultado = (media**x)*(e**(-media))/fatorial(x)
    return resultado
```

## Distribuição de probabilidade comum:

### Time mandante:

```python
#agrupamento
a = df.groupby(['mandante', 'placar_m'])['placar_m'].count().unstack()

#tratando dados faltantes (vai ser util para um monte de coisa)
n = df['placar_m'].unique()

for i in n:
    a[i].fillna(0, inplace = True)
    
#criando coluna com o total para tirar as metricas
a['total'] = 0

for i in n:
    a['total'] += a[i]
    
#criando coluna com a media de gols
a['soma ponderada'] = 0

for i in n:
    a['soma ponderada'] += a[i]*i

a['media de gols'] = a['soma ponderada']/a['total']

b = a.copy()

#transformando em probabilidade - distribuição de probabilidade
for i in n:
    a[i] = a[i]/a['total']
    
#tirando as colunas que não mais preciso
a.drop(['total', 'soma ponderada'], 1, inplace=True)
a

#deixando em porcentagem para melhor visualização
a_per = a.copy()

for i in n:
    a_per[i] = a_per[i].apply(lambda x: x*100)
    
a_per
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

a_per.T.plot(figsize=(20,12))
```

### Time visitante:

```python
#agrupamento
c = df.groupby(['visitante', 'placar_v'])['placar_v'].count().unstack()

#tratando dados faltantes (vai ser util para um monte de coisa)
n = df['placar_v'].unique()

for i in n:
    c[i].fillna(0, inplace = True)
    
#criando coluna com o total para tirar as metricas
c['total'] = 0

for i in n:
    c['total'] += c[i]
    
#criando coluna com a media de gols
c['soma ponderada'] = 0

for i in n:
    c['soma ponderada'] += c[i]*i

c['media de gols'] = c['soma ponderada']/c['total']

d = c.copy()

#transformando em probabilidade - distribuição de probabilidade
for i in n:
    c[i] = c[i]/c['total']
    
#tirando as colunas que não mais preciso
c.drop(['total', 'soma ponderada'], 1, inplace=True)

#deixando em porcentagem para melhor visualização
c_per = c.copy()

for i in n:
    c_per[i] = c_per[i].apply(lambda x: x*100)
    
c_per
```

```python
c_per.T.plot(figsize=(20,12))
```
