```python
import requests
from bs4 import BeautifulSoup

linkJogo = 'https://www.espn.com.br/futebol/resultados/_/liga/BRA.1/data/20211209'
requisicao = requests.get(linkJogo)
soup = BeautifulSoup(requisicao.content, 'html.parser')
soup
```

```python
for i in soup.find_all(class_='scoreboard soccer fallback js-show'):
    print(i.get_text())
```

```python
for i in soup.find_all(class_ = 'scoreboard soccer fallback js-show'):
    print(i)
```

```python
soup.find(class_ = 'scoreboard soccer fallback js-show')
```

```python
'''CA'''
texto = "/futebol/partida-estatisticas/"
lista = []

for i in soup.find_all("events"):
    var = i.get('href')
    if var[0:30] == texto:
        lista.append(var)
        
from collections import OrderedDict
lista = OrderedDict.fromkeys(lista)

print('tamanho da lista:', len(lista))
list(lista)
```

```python
#Times
times = []
for i in soup.find_all(class_ = 'team-name'):
    times.append(i.get_text())
    
    
count = 0
mandante = []
visitante = []

for i in times:
    count += 1
    if count % 2 == 1:
        mandante.append(i)
    else:
        visitante.append(i)
        
print(mandante)
print(visitante)
```

```python
placar = []
#Resultados dos jogos
for i in soup.find_all(class_ = 'record'):
    placar.append(i.get_text())
    
print(placar)    
```

```python
link = 'https://www.espn.com.br'

jogo = link+base['link'][3]

requisicao = requests.get(jogo)
sp = BeautifulSoup(requisicao.content, 'html.parser')
sp
```

```python
#Posse da bola
Possebola = []
for i in sp.find_all('span', attrs={"class": "chartValue"}):
    Possebola.append(i.get_text())
    
'''OUTRAS ESTATISTICAS'''
estatisticas = sp.find('div', attrs={"class": "stat-list"})

#Faltas
Faltas = []
for i in estatisticas.find_all('td', attrs={'data-stat':'foulsCommitted'}):
    Faltas.append(i.get_text())
    
#Cartões Amarelos
CartõesAmarelos = []
for i in estatisticas.find_all('td', attrs={'data-stat':'yellowCards'}):
    CartõesAmarelos.append(i.get_text())
    
#Cartões Vermelhos
CartõesVermelhos = []
for i in estatisticas.find_all('td', attrs={'data-stat':'redCards'}):
    CartõesVermelhos.append(i.get_text())

#Impedimentos
Impedimentos = []
for i in estatisticas.find_all('td', attrs={'data-stat':'offsides'}):
    Impedimentos.append(i.get_text())

#Escanteios
Escanteios = []
for i in estatisticas.find_all('td', attrs={'data-stat':'wonCorners'}):
    Escanteios.append(i.get_text())

#Defesas
Defesas = []
for i in estatisticas.find_all('td', attrs={'data-stat':'saves'}):
    Defesas.append(i.get_text())


'''VERIFICANDO VALORES'''
print(Possebola)
print(Faltas)
print(CartõesAmarelos)
print(CartõesVermelhos)
print(Impedimentos)
print(Escanteios)
print(Defesas)
```
