# Identificando a localização das informações requeridas na pagina URL

```python
import requests
from bs4 import BeautifulSoup

linkJogo = "https://onefootball.com//pt-br/match/2205040"
requisicao = requests.get(linkJogo)
soup = BeautifulSoup(requisicao.content, 'html.parser')
soup
```

```python
#mandante
print(soup.find(class_="match-score-team__name title-7-medium match-score-team__name--home").get_text())
#visitante
print(soup.find(class_="match-score-team__name title-7-medium match-score-team__name--away").get_text())
#placar
print(soup.find(class_="match-score-scores").get_text())
```

```python
'''----Estatisticas----
    Posse de bola
    Total de chutes
    Chutes ao gol
    Disputas ganhas'''
#Estatisticas mandante
print('-'*10, 'mandante:', '-'*10)
for i in soup.find_all(class_="title-7-bold match-stats__stat-value"):
    print(i.get_text())
    
#Estatisticas visitante
print('-'*10, 'visitante:', '-'*10)
for i in soup.find_all(class_="title-7-bold match-stats__stat-value match-stats__stat-value--away"):
    print(i.get_text())
```

```python
'''Campeonato
    Data do jogo
    Arena'''
#outros dados
for i in soup.find_all(class_="title-8-regular match-info__entry-subtitle"):
    print(i.get_text())
```

# Capturando as informações de todos os jogos

```python
#teste = df.iloc[1:6]
#teste = df.sample(15)

#Criando listas para as variaveis
m = []
v = []
placar = []
est_m = []
est_v = []
outros = []
    
count = 0
for i in df['links']:
    count += 1
    linkJogo = "https://onefootball.com"+i
    print(count, "-"*10, linkJogo, "-"*10)
    requisicao = requests.get(linkJogo)
    soup = BeautifulSoup(requisicao.content, 'html.parser')

    '''CAPTURA'''
    m.append(soup.find(class_="match-score-team__name title-7-medium match-score-team__name--home").get_text())
    v.append(soup.find(class_="match-score-team__name title-7-medium match-score-team__name--away").get_text())
    placar.append(soup.find(class_="match-score-scores").get_text())
    
    #Estatisticas mandante
    est_1 = []
    for t in soup.find_all(class_="title-7-bold match-stats__stat-value"):
        est_1.append(t.get_text())
    est_m.append(est_1)
        
    #Estatisticas visitante
    est_2 = []
    for tt in soup.find_all(class_="title-7-bold match-stats__stat-value match-stats__stat-value--away"):
        est_2.append(tt.get_text())
    est_v.append(est_2)
        
    #outros dados
    out = []
    for ttt in soup.find_all(class_="title-8-regular match-info__entry-subtitle"):
        out.append(ttt.get_text())
    outros.append(out)
    
    print("OK.")
```

```python
'''------VERIFICAÇÃO------'''
print('m:', m, '\n')
print('v:', v, '\n')
print('placar:', placar, '\n')
print('est_m:', est_m, '\n')
print('est_v:', est_v, '\n')
print('outros:', outros, '\n')
```

```python
'''------ORGANIZANDO TUDO EM UM DATAFRAME------'''
base = pd.DataFrame({'mandante': m ,
                       'placar': placar,
                      'visitante': v})
'''
p = posse de bola
tc = total de chutes
cg = chutes ao gol
d = disputas ganhas
'''
#mandante
tabela = pd.DataFrame(est_m)
tabela.rename(columns={0:'p_m', 1:'tc_m', 2:'cg_m', 3:'d_m'}, inplace = True) #***
base = base.join(tabela)
#visitante
tabela = pd.DataFrame(est_v)
tabela.rename(columns={0:'p_v', 1:'tc_v', 2:'cg_v', 3:'d_v'}, inplace = True) #***
base = base.join(tabela)

#outros dados
tabela = pd.DataFrame(outros)
tabela.rename(columns={0:'campeonato', 1:'data', 2:'local'}, inplace = True) #***
base = base.join(tabela)

base
```

# PRÉ PROCESSAMENTOS DOS DADOS

```python
'''------CONCERTANDO FORMATO DOS DADOS------'''

#transformando em formato data
base['data'] = pd.to_datetime(base['data'])

#separando os placares
base['placar_m'] = base['placar'].str[0]
base['placar_v'] = base['placar'].str[-1]

base = base.drop(columns=['placar'])#exclui a coluna placar fonte

#Juntando cada qual com seu link
base = base.join(df)
base
```

```python
'''------ALTERANDO ORDEM DOS DADOS NO DATAFRAME------'''

df = base[['data', 'mandante', 'placar_m', 'placar_v', 'visitante',
      'p_m', 'p_v', 'tc_m', 'tc_v', 'cg_m', 'cg_v', 'd_m', 'd_v',
      'local', 'campeonato', 'links']]

df
```

```python
#Nota-se no processo de extração que algumas informações podiam vim com espaços a mais a direita ou esquerda dos dados
#Retirando espaços
for i in df.columns:
    try:
        df[i] = df[i].apply(lambda x: x.strip())
    except:
        continue 
```

```python
#Transformando em porcentagem
lista = ['p_m', 'p_v', 'd_m', 'd_v']

for i in lista:
    df[i] = df[i].apply(lambda x: int(x[:-1])/100)
df
```

```python
#Salvando a tabela
df.to_csv("jogo_bras.csv", index=False)
```
