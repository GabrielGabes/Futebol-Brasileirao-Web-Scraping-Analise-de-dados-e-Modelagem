```python
import requests
from bs4 import BeautifulSoup

linkJogo = 'https://www.espn.com.br/futebol/fixtures/_/data/20210930'
requisicao = requests.get(linkJogo)
soup = BeautifulSoup(requisicao.content, 'html.parser')
soup
```

# Localizando os atributos requeridos

```python
#Campeonatos
for campeonato in soup.find_all('h2', class_ = 'table-caption'):
    print(campeonato.get_text())
```

```python
#Times
for time in soup.find_all(class_ = 'team-name'):
    print(time.get_text())
```

```python
texto = "/futebol/partida-estatisticas/"
lista = []

for i in soup.find_all("a"):
    var = i.get('href')
    if var[0:30] == texto:
        lista.append(var)
        
lista
```

```python
from collections import OrderedDict
lista_final = OrderedDict.fromkeys(lista)
```

```python
len(lista_final)
```

```python
#Resultados dos jogos
for resultado in soup.find_all(class_ = 'record'):
    print(resultado.get_text())
```

```python
#Local do jogo
lista = []
for local in soup.find_all(class_ = 'schedule-location'):
    print(local.get_text())
    lista.append(local.get_text())
```

```python
len(lista)
```

```python
#Tudo junto
for jogo in soup.find_all(attrs={'class' : 'responsive-table-wrap'}):
        print('>',jogo.get_text())
```

```python
todos_jogos = []

for i in soup.find_all(class_ = 'odd has-results'):
    todos_jogos.append(i)
    
for i in soup.find_all(class_ = 'even has-results'):
    todos_jogos.append(i)
```

```python
lista = []

for divisao in todos_jogos:
    #campeonato = divisao.find(class_ = 'table-caption').get_text()
    local = divisao.find(class_ = 'schedule-location').get_text()
    resultado = divisao.find(class_ = 'record').get_text()
    lista.append((resultado, local))
    
print(lista)
```
