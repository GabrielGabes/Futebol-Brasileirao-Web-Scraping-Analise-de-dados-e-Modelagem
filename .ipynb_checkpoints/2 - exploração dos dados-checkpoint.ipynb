```python
import pandas as pd

#Carregando os dados
df = pd.read_csv("jogo_bras.csv")

#ADICIONANDO COLUNA QUE INDICA O VENCEDOR
vencedor = []
derrotado = []
count = 0
for i in df['data']:
    #print(count)
    if df['placar_m'][count] > df['placar_v'][count]:
        vencedor.append(df['mandante'][count])
        derrotado.append(df['visitante'][count])
    elif df['placar_m'][count] < df['placar_v'][count]:
        vencedor.append(df['visitante'][count])
        derrotado.append(df['mandante'][count])
    else:
        vencedor.append('-')
        derrotado.append('-')
    count += 1
    
df['vencedor'] = vencedor
df['derrotado'] = derrotado
df
```

```python
'''QUANTIDADE DE VITORIAS E DERROTAS DE CADA TIME'''
a = df['vencedor'].value_counts().reset_index()
a['perc'] = a['vencedor']*100/38
display(a)
b = df['derrotado'].value_counts().reset_index()
b['perc'] = b['derrotado']*100/38
display(b)
```
