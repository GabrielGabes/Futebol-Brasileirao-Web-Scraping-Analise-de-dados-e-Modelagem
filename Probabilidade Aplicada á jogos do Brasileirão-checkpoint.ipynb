```python
#Fatorial de um numero
def fatorial(numero):
    resultado=1
    count=1

    while count <= numero:
        resultado *= count
        count += 1
        
    return resultado

#Probabilidade Poisson
def dist_poisson(media, x):
    resultado = (media**x)*(e**(-media))/fatorial(x)
    return resultado
```

## -Times como Mandante

## Distribuição de Probabilidade

```python
#agrupamento
a = test.groupby(['mandante', 'mandante placar'])['mandante placar'].count().unstack()

#tratando dados faltantes (vai ser util para um monte de coisa)
n = test['mandante placar'].unique()

for i in n:
    a[i].fillna(0, inplace = True)
    
#criando coluna com o total para tirar as metricas
a['total'] = 0

for i in n:
    a['total'] += a[i]
    
#criando coluna com a media de gols
a['soma ponderada'] = 0

for i in n:
    a['soma ponderada'] += a[i]*i

a['media de gols'] = a['soma ponderada']/a['total']

b = a.copy()

#transformando em probabilidade - distribuição de probabilidade
for i in n:
    a[i] = a[i]/a['total']
    
#tirando as colunas que não mais preciso
a.drop(['total', 'soma ponderada'], 1, inplace=True)
a

#deixando em porcentagem para melhor visualização
a_per = a.copy()

for i in n:
    a_per[i] = a_per[i].apply(lambda x: x*100)
    
a_per
```

## Distribuição Poisson

```python
#aplicando formula em cada valor
for i in n:
    b[i] = dist_poisson(b['media de gols'], i)

#tirando as colunas que não mais preciso
b.drop(['total', 'soma ponderada'], 1, inplace=True)

#deixando em porcentagem para melhor visualização
b_per = b.copy()

for i in n:
    b_per[i] = b_per[i].apply(lambda x: x*100)
    
b_per
```

# Probabilidade vitoria mandante

```python
def prob_elo(elo_mandante, elo_visitante):
    prob = (elo_mandante/400)/((elo_mandante/400)+(elo_visitante/400)) #probabilidade de vitoria do time mandante
    return print('A probabilidade do time mandante ganhar do time visitante baseando-se no elo de ambos é de:', prob*100)
```
