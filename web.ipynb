```python
incremento = 'https://www.espn.com.br/futebol/partida-estatisticas/_/jogoId/'

#Listas com os 
campeonato = []

abrevitura_v = []
abrevitura_m = []
v = []
m = []
placar_v = []
placar_m = []

Possebola_v = []
Faltas_v = []
CartõesAmarelos_v = []
CartõesVermelhos_v = []
Impedimentos_v = []
Escanteios_v = []
Defesas_v = []
Possebola_m = []
Faltas_m = []
CartõesAmarelos_m = []
CartõesVermelhos_m = []
Impedimentos_m = []
Escanteios_m = []
Defesas_m = []

for i in links_jogos:
    partida = incremento+i
    site = requests.get(partida)
    sp = BeautifulSoup(site.content, 'html.parser')
    
    #Campeonato
    campeonato.append(sp.find(class_ = "game-details header").get_text(strip=True))
    
    visiting = sp.find(class_ = 'team__content')
    home = sp.find(class_ = 'team home')

    #Abreviatura do nome dos times
    #abrevitura_v.append(visiting.find(class_ = 'abbrev').get_text())
    #abrevitura_m.append(home.find(class_ = 'abbrev').get_text())

    #Nome dos times
    v.append(visiting.find(class_ = 'long-name').get_text())
    m.append(home.find(class_ = 'long-name').get_text())

    #Placar
    placar_v.append(visiting.find(class_ = 'score icon-font-after').get_text(strip=True))
    placar_m.append(home.find(class_ = 'score icon-font-before').get_text(strip=True))
    
    #Posse da bola
    Possebola = []
    for i in sp.find_all('span', attrs={"class": "chartValue"}):
        Possebola.append(i.get_text())

    '''OUTRAS ESTATISTICAS'''
    estatisticas = sp.find('div', attrs={"class": "stat-list"})

    #Faltas
    Faltas = []
    for i in estatisticas.find_all('td', attrs={'data-stat':'foulsCommitted'}):
        Faltas.append(i.get_text())

    #Cartões Amarelos
    CartõesAmarelos = []
    for i in estatisticas.find_all('td', attrs={'data-stat':'yellowCards'}):
        CartõesAmarelos.append(i.get_text())

    #Cartões Vermelhos
    CartõesVermelhos = []
    for i in estatisticas.find_all('td', attrs={'data-stat':'redCards'}):
        CartõesVermelhos.append(i.get_text())

    #Impedimentos
    Impedimentos = []
    for i in estatisticas.find_all('td', attrs={'data-stat':'offsides'}):
        Impedimentos.append(i.get_text())

    #Escanteios
    Escanteios = []
    for i in estatisticas.find_all('td', attrs={'data-stat':'wonCorners'}):
        Escanteios.append(i.get_text())

    #Defesas
    Defesas = []
    for i in estatisticas.find_all('td', attrs={'data-stat':'saves'}):
        Defesas.append(i.get_text())

    #Colocando tudo em seu devido lugar
    count = 0
    for i in range(0,2):
      count += 1
      if count % 2 == 1:
        Possebola_v.append(Possebola[i])
        Faltas_v.append(Faltas[i])
        CartõesAmarelos_v.append(CartõesAmarelos[i])
        CartõesVermelhos_v.append(CartõesVermelhos[i])
        Impedimentos_v.append(Impedimentos[i])
        Escanteios_v.append(Escanteios[i])
        Defesas_v.append(Defesas[i])
      else:
        Possebola_m.append(Possebola[i])
        Faltas_m.append(Faltas[i])
        CartõesAmarelos_m.append(CartõesAmarelos[i])
        CartõesVermelhos_m.append(CartõesVermelhos[i])
        Impedimentos_m.append(Impedimentos[i])
        Escanteios_m.append(Escanteios[i])
        Defesas_m.append(Defesas[i])    
```
